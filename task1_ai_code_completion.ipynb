```python
# ✅ Task 1: AI-Powered Code Completion
# Objective: Sort a list of dictionaries by a specific key

# Sample data
students = [
    {"name": "Alice", "score": 85},
    {"name": "Bob", "score": 92},
    {"name": "Charlie", "score": 78},
    {"name": "David", "score": 90}
]

# ----------------------------
# Manual Implementation
# ----------------------------
def sort_dicts_manual(data, key):
    """
    Sorts a list of dictionaries by a specific key manually.
    """
    return sorted(data, key=lambda x: x[key])

manual_sorted = sort_dicts_manual(students, "score")
print("Manual sorted result:")
print(manual_sorted)

# ----------------------------
# AI-Suggested Implementation
# (Generated by GitHub Copilot / Tabnine)
# ----------------------------
def sort_dicts_ai(data, key):
    """
    AI-assisted sorting of list of dictionaries by a given key.
    """
    return sorted(data, key=lambda item: item[key])

ai_sorted = sort_dicts_ai(students, "score")
print("\nAI-assisted sorted result:")
print(ai_sorted)

# ---------------------------
#Word Analysis
# ----------------------------

#The comparison between manual implementation and AI-assisted code completion shows that AI tools
#like GitHub Copilot or Tabnine can generate efficient and accurate code almost instantly. Both
#versions produce identical results for sorting a list of dictionaries by a specific key. From a
#performance standpoint, there is negligible difference because both implementations rely on Python’s
#built-in `sorted()` function, which uses Timsort and has a time complexity of O(n log n).

#The main advantage of AI-assisted coding lies in time efficiency and developer productivity. With AI,
#developers can focus on higher-level logic and design, while repetitive or boilerplate code is
#generated automatically. This also reduces the risk of syntax errors or minor bugs. AI tools can
#provide suggestions for edge cases, documentation, and even optimize code readability, which might
#be overlooked in manual implementation.

#However, manual implementation encourages deeper understanding of Python concepts, including lambda
#functions, key-based sorting, and list comprehension. While AI accelerates coding, it may not teach
#est practices unless the developer reviews the suggestions carefully. In summary, AI-assisted code
#completion is an invaluable tool for improving efficiency, reducing errors, and speeding up
#development cycles, particularly in repetitive or standard coding tasks.

```
